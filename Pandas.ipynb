```python
carsales = new_data.copy()
carsales
```

```python
seat_column = pd.Series([5,5,5,5,5])
carsales["Seats"] = seat_column
carsales
```

```python
carsales['Seats'].fillna(5.0, inplace=True)
```

```python
carsales
```

```python
fuel_economy = [3.2,6.1,3.2,1.8,2.0,7.3,8.12,9.0,7.5,4.3]
carsales['Fuel per 100km'] = fuel_economy
carsales
```

```python
carsales['Fuel usage'] = carsales['Odometer (KM)']/100 * carsales['Fuel per 100km'
                                                                 ]
carsales
```

```python
carsales['Passes Test'] = carsales['Odometer (KM)']>50000
carsales
```

```python
carsales['Odometer (KM)'] = carsales['Odometer (KM)'].apply(lambda x: x/1.6)
carsales
```

```python
carsales.rename(columns={'Odometer (KM)': 'Odometer(Miles)'})
```

```python
carsales.to_csv('carsales.csv', index=False)
```
